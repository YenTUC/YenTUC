## Iteration
- Iterable: an object that you can iterate over (e.g. a list), with a `for` loop.
- Container: an object that contains other objects, you can use `in` on it.
- Element, item: an object that is contained by a container or produced by iterating on an iterable.
- Sequence: an ordered container.
- Iterator: an object that provides an iteration over an iterable; one iterable can have several, independent, non-identical iterators.
- Comprehensions: a compact way to process all or part of the elements in an iterable and return a new container.

```python
for i in 0.1,-1,0,3:  #for loop
    print(i)
```

```python
import numpy as np
for i in range(1,10,2):   # for loop with range
    x = i/2
    rounding = np.round(x)
    print(i, x, rounding)
```

```python
from random import randint 
random_num = randint(1,100)
while random_num <= 70:         # condition
    print(random_num)           
    random_num = randint(1,100) 
print ('Found a number greater than 70:', random_num)
```

## Try and except
You can catch an error using a try-except.

```python
filename = "datatable22.csv"
try:
    with open(filename) as f:
        print(f.read())
except FileNotFoundError:
    print("File", filename, "not found, please try a different filename")
```

EAFP - easier to ask forgiveness than permission:

```python
number = input('Give me a number please: ')
try:
    number = int(number)
except ValueError:
    print("I asked for a number and you gave me:", number, " - this can't be parsed to int")
```

## Date and Time

```python
from datetime import datetime
now = datetime.now()
now
```

You can format time using strftime:

```python
now.strftime("%H:%M %d.%m.%Y")
```

You can use strptime to parse time from strings:

```python
dt = datetime.strptime('21:50 14.03.2017', "%H:%M %d.%m.%Y")
dt
```

## Timing a program's runtime

```python
%timeit y = [x**2 for x in range(10000000)] # list comprehension
```

```python
%%timeit
y = []
for x in range(1000000):
    y.append(x**2)
for x in range(1000000):
    y.append(x**2)
```
